```python
import pycountry
import plotly.express as px

# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2 = code).alpha_3
  except:
    return None

# Criar nova coluna com código ISO-3
df['residencia_iso3'] = df['residencia'].apply(iso2_to_iso3)

# Calcular média salarial por país (ISO-3)
df_ds = df[df['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['salario_usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations = 'residencia_iso3',
                    color = 'salario_usd',
                    color_continuous_scale = 'rdylgn',
                    title = 'Salário médio de Cientista de Dados por País',
                    labels = {'salario_usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
```

```python
import plotly.express as px
```

```python
senioridade_media_salario = df.groupby('senioridade')['salario_usd'].mean().round(2).sort_values(ascending = False).reset_index()
```

```python
fig = px.bar(senioridade_media_salario,
             x = 'senioridade',
             y = 'salario_usd',
             title = 'Média Salarial por Senioridade',
             labels = {'senioridade':'Nível de senioridade', 'salario_usd': 'Média Salarial Anual (USD)'})

fig.show()
```

```python
# Desafio
# Entender Salário Médio pelo País que a pessoa está, com o cargo de Cientista de Dados. Usando plotly.
```

```python
df["cargo"].unique()
```

```python
salario_medio_pais = (
    df[df["cargo"] == "Data Scientist"]
    .groupby("residencia")["salario_usd"]
    .mean()
    .reset_index()
    .sort_values("salario_usd", ascending=False)
)
```

```python
import plotly.express as px

fig = px.bar(
    salario_medio_pais,
    x="residencia",
    y="salario_usd",
    title="Salário médio por país — Cientista de Dados",
    labels={
        "residencia": "País de Residência",
        "salario_usd": "Salário Médio Anual (USD)"
    }
)

fig.show()
```
